```python
# Importing some packages
import matplotlib.pyplot as plt
import numpy as np

# Initializing x and y
x=np.arange(2010,2090,5)
y=np.random.randint(0,100,size=x.shape)

# Common line style as dictionary
line_style=dict(fontweight="bold", fontsize=10,family="Serif")

# Common maker style as dictionary
marker_style=dict(marker=".", markersize=10, markerfacecolor="red", markeredgecolor="red",linestyle="None")

# Grid style as dictionary
grid_style=dict(axis="both", color="yellow", linestyle="dashed", linewidth=2)

#Setting the title 
plt.title("Gold Price Over the passed 100 years",color="yellow",**line_style)

#Setting X axis
plt.xlabel("Year",color="red",**line_style)
#Setting Y axis
plt.ylabel("Price",color="green",**line_style)

#Calling the grid method
plt.grid(**grid_style)

#Plotting the graph
plt.plot(x,y,**marker_style)

#Showing the graph onto the console
plt.show()
```
