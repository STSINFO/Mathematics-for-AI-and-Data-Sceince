```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f_example_2(x,y):
    # return x**4+0.8*y**4+4*x**2+2*y**2-x*y-0.2*x**2*y
     return x**4 + 0.8*y**4 + 4*x**2 + 2*y**2 - x*y -0.2*x**2*y
def f_grad_example_2(x,y):
    # return np.array([[4*x**3+8*x**3-y-0.4*x*y],[3.2*y**3+4*y-0.2*x**2]])
    return np.array([[4*x**3 + 8*x - y - 0.4*x*y],
                     [3.2*y**3 +4*y - x - 0.2*x**2]])

def f_hessian_example_2(x,y):
    # return np.array([[12*x**2+24*x-0.4, -1-0.4*x],[-1-0.4*x, 9.6*y**2-4]])
    return np.array([[12*x**2 + 8 - 0.4*y, -1 - 0.4*x],
                         [-1 - 0.4*x, 9.6*y**2 + 4]])

def gradient_decent_2(grad,x_y,learning_rate=0.01,num_iterations=25):
    for iteration in range(num_iterations):
        x_y=x_y-learning_rate*grad(x_y[0,0],x_y[1,0])
        print(x_y.T)
    return x_y

def newton_method_2(grad,hessian,x_y,num_iterations=25):
    for iteration in range(num_iterations):
        x_y=x_y-np.matmul(np.linalg.inv(hessian(x_y[0,0],x_y[1,0])),grad(x_y[0,0],x_y[1,0]))
        print(x_y.T)
    return x_y    
```

```python
x_y=np.array([[4],[4]])

```

```python
print(f_example_2(x_y[0,0],x_y[1,0]))
print(f_grad_example_2(x_y[0,0],x_y[1,0]))
print(f_hessian_example_2(x_y[0,0],x_y[1,0]))
```

```python
gradient_decent_2(f_grad_example_2,x_y)
```
